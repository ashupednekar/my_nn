```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading dataset
data = pd.read_csv('tagged_data.csv')
```

```python
data.head()
```

```python
#featurising the inputs into non-string values
def get_unique_word(df):
    words = list(set(df.word))
    unique = dict()
    for i in range(len(words)):
        unique[words[i]] = i
    return unique

def get_unique_tag(df):
    words = list(set(df.tag))
    unique = dict()
    for i in range(len(words)):
        unique[words[i]] = i
    return unique

u = get_unique_word(data)
t = get_unique_tag(data)

# print(u)
# print(t)

# labels, _ = pd.factorize(data.word)
u
```

```python
X = np.array([u[word] for word in data.word]).reshape(-1, 1)
y = np.array([t[tag] for tag in data.tag]).reshape(-1, 1)
```

```python
# from sklearn.naive_bayes import GaussianNB
# clf = GaussianNB()

from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
#training
clf.fit(X, y)
```
