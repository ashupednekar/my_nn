```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return (x * (1 - x))
```

```python
#First and foremost, I've created the basic class to represent each nodeof the neural network arhiteture.
class NeuralNet:
    #constructor to initializw
    def __init__(self, x, y):
        self.input = x
        self.weights1 = np.random.rand(self.input.shape[1],1) #since this network consists of only one hidden layer with a single node.
        self.weights2 = np.random.rand(1,1)                 
        self.y = y
        self.output = np.zeros(y.shape)
        
    #For the forward pass, I created this function.
    #All it does is calculates the weighted sum of the weights and the features and then passes it through the sigmoid function.
    def forward_pass(self):
        self.layer1 = sigmoid(np.dot(self.input, self.weights1))
        self.output = sigmoid(np.dot(self.layer1, self.weights2))
        return self.output
    
    def backward_pass(self):
        #Using the formula for 
        d_weights2 = np.dot(self.layer1.T, (2*(self.y - self.output) * sigmoid_derivative(self.output)))
        d_weights1 = np.dot(self.input.T,  (np.dot(2*(self.y - self.output) * sigmoid_derivative(self.output), self.weights2.T) * sigmoid_derivative(self.layer1)))

        # update the weights with the derivative (slope) of the loss function
        self.weights1 += d_weights1
        self.weights2 += d_weights2
```

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

```python
network = NeuralNet(X,y)
output = network.forward_pass()
#initial output before training
output
```

```python
#training
for i in range(10000):
    network.backward_pass()
```
